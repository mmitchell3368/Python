````python
# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['SALES'], bins=30, stat="density", color='skyblue')
sns.kdeplot(df['SALES'], color="orange")
plt.title('Histogram of SALES')
plt.xlabel('SALES')
plt.ylabel('Frequency')
plt.show()
````

````python
def get_mad(s):
    median = np.median(s)
    diff = np.abs(s-median)
    MAD = np.median(diff)
    return MAD
````

````python
MAD = get_mad(df['SALES'])
MAD
````

````python
median = np.median(df['SALES'])
median
````

````python
def get_modified_z_score(x, median, MAD):
    return 0.6745*(x-median)/MAD
````

````python
df['mod_z_score'] = df.SALES.apply(lambda x: get_modified_z_score(x, median, MAD))
df.head()
````

````python
df_new = df[(df.mod_z_score < 3.5) & (df.mod_z_score > -3.5)]
````

````python
df_new
````

````python
# Modified Histogram
plt.figure(figsize=(10, 6))
sns.histplot(df_new['SALES'], bins=30, stat="density", color='skyblue')
sns.kdeplot(df_new['SALES'], color="orange")
plt.title('Histogram of SALES')
plt.xlabel('SALES')
plt.ylabel('Frequency')
plt.show()
````

````python
# Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='STATUS', y='SALES', data=df, hue='STATUS', legend=False)
plt.title('Boxplot of SALES by STATUS')
plt.xlabel('STATUS')
plt.ylabel('SALES')
plt.show()
````

````python
# KDE plot
plt.style.use("seaborn-v0_8-bright")
plt.figure(figsize=(10, 6))
sns.kdeplot(df['QUANTITYORDERED'], fill=True, color="orange")
plt.title('KDE Plot of QUANTITYORDERED')
plt.xlabel('QUANTITYORDERED')
plt.ylabel('Density')
plt.show()
````

# Step 4: Feature Relationships 
- Scatterplot 
- Heatmap Correlation 
- Pairplot 
- Groupby comparisons

````python
df_use = df.drop(['mod_z_score'], axis=1)
````

````python
df_use.head()
````

````python
df_use.columns
````

````python
# Scatter plot
plt.scatter(df_use['QUANTITYORDERED'], df_use['SALES'])
plt.xlabel('Quantity Ordered')
plt.ylabel('Sales')
plt.title('Scatter Plot: Quantity Ordered vs Sales')
plt.show()
````

````python
# Heatmap
correlation_matrix = df_use.corr(numeric_only=True)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
````

````python
# Pair plot
sns.pairplot(df[['QUANTITYORDERED', 'PRICEEACH', 'SALES', 'DAYS_SINCE_LASTORDER']])
plt.suptitle('Pair Plot', y=1.02)
plt.show()
````

````python
# Groupby comparisons
grouped_data = df_use.groupby('PRODUCTLINE')['SALES'].mean().reset_index()
grouped_data
````

````python
# Bar plot of average sales per product line
plt.figure(figsize=(12, 6))
sns.barplot(x=grouped_data['PRODUCTLINE'], y=grouped_data['SALES'])
plt.xlabel('Product Line')
plt.ylabel('Average Sales')
plt.title('Average Sales per Product Line')
plt.xticks(rotation=45)
plt.show()
````

````python
# Box plot of sales distribution per product line
plt.figure(figsize=(12, 6))
sns.boxplot(x=df_use['PRODUCTLINE'], y=df['SALES'])
plt.xlabel('Product Line')
plt.ylabel('Sales')
plt.title('Sales Distribution per Product Line')
plt.xticks(rotation=45)
plt.show()
````

# Step 5: Ask a Question about the Data
- Try to answer a question you have about the data using a plot or statistic

- Question: Which customer spend the most money on each product?

````python
df_use.head()
````

````python
df_use.columns
````

````python
# Group by 'PRODUCTCODE' and 'CUSTOMERNAME', summing up the 'SALES' for each combination
grouped_data = df_use.groupby(['PRODUCTLINE', 'CUSTOMERNAME'])['SALES'].sum().reset_index()
````

````python
# Find the customer with the maximum total sales for each product
max_spenders = grouped_data.groupby('PRODUCTLINE')[['CUSTOMERNAME', 'SALES']].max()
````

````python
max_spenders
````

````python
# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='PRODUCTLINE', y='SALES', hue='CUSTOMERNAME', data=max_spenders)
plt.xlabel('Product Line')
plt.ylabel('Sales')
plt.title('Sales by Product Line and Customer')
plt.legend(title='Customer Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.show()
````
