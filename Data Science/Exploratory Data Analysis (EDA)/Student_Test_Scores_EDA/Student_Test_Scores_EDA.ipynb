# Step 2: Cleaning the data

```python
data.isnull().sum()
```

```python
student = data.drop(['race/ethnicity', 'parental level of education'], axis = 1)
```

```python
student.head()
```

# Step 3: Relationship Analysis

```python
corr = student.corr(numeric_only=True)
```

```python
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns,
           annot=True)
```

```python
sns.pairplot(student)
```

```python
sns.relplot(student, x='math score', y='reading score', hue='gender')
```

```python
sns.relplot(student, x='math score', y='reading score', hue='lunch')
```

```python
sns.displot(student['math score'], bins=10, kde=True)
```

```python
sns.catplot(x='reading score', kind='box', data=student)
```
